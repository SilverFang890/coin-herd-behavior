```python
"""
put the txt files into datascience/pandas table datatype

"""
# Run this cell, but please don't change it.

# These lines import the Numpy and Datascience modules.
import numpy as np

# These lines do some fancy plotting magic.
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')
import warnings
warnings.simplefilter('ignore', FutureWarning)
warnings.simplefilter('ignore', UserWarning)

from pycoingecko import CoinGeckoAPI
from datascience import *
import time
cg = CoinGeckoAPI()
```

```python
memecoins = ['100x-coin', 'akita-inu', 'chow-chow-finance', 'cumrocket', 'dogecoin', 'dogelon-mars', 
             'elongate', 'husky', 'memepad', 'memes-token', 'monacoin', 'pepemon-pepeballs', 'pig-finance', 
             'poocoin', 'popcorn-token', 'rotten', 'sad-cat-token', 'shiba-inu', 'tendies', 'safemoon', 'jejudoge',
             'smooth-love-potion']

prices_list = []
for i in range(len(memecoins)):
    prices_list.append([])


counter = 0

for coin in memecoins:
    with open(f'./memecoins/{coin}.txt', 'r') as coin_handle:
        coin_array = coin_handle.readlines()
        coinstring = coin_array[0]
        coin_price_array = coinstring.split(',')
        for price in coin_price_array:
            price = float(price)
            prices_list[counter].append(price)
        counter += 1

        
```
